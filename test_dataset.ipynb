```python
from datasets import load_dataset

from PIL import Image, ImageTk
import matplotlib.pyplot as plt
import random

# Login using e.g. `huggingface-cli login` to access this dataset
dataset = load_dataset("ColumbiaNLP/V-FLUTE")

# Get a random index in the training set
random_index = random.randint(0, len(dataset['train']) - 1)

# Print a random example
firstImage = dataset['train'][random_index].get("image")
firstClaim = dataset['train'][random_index].get("claim")
firstLabel = dataset['train'][random_index].get("label")
firstExplanation = dataset['train'][random_index].get("explanation")
firstPhenomenon = dataset['train'][random_index].get("phenomenon")
firstSourceDataset = dataset['train'][random_index].get("source_dataset")


#print the output
print("Label: " + firstLabel)
print("Explanation: " + firstExplanation)
print("Phenomenon: " + firstPhenomenon)
print("SourceDataset: " + firstSourceDataset)


# Display the image and claim using Matplotlib
plt.imshow(firstImage)
plt.title("Claim: " + firstClaim)

plt.axis('off')  # Hide axes
plt.show()
```
